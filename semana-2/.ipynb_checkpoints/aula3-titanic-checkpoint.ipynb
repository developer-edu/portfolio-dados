```python
import pandas as pd
df_titanic = pd.read_csv('Titanic-Dataset.csv')
```

## Manipulação de DataFrames com Pandas
Selecionar, fatiar e filtrar dados de um DataFrame para responder perguntas específicas.

```python
# Selecionar a coluna 'Age' (idade)
idades = df_titanic['Age']
print(type(idades))
idades.head() # Usamos .head() porque é uma Series longa
```

```python
# Selecionar as colunas 'Name', 'Sex' e 'Age'
# Note os colchetes duplos: df[[...]]
df_subset = df_titanic[['Name', 'Sex', 'Age']]
df_subset.head()
```

```python
## Selecionando Linhas com .iloc (seleção por Posição)
# Selecionar a PRIMEIRA linha do DataFrame
primeira_linha = df_titanic.iloc[0]
print(primeira_linha)

# Selecionar as 3 primeiras linhas
tres_primeiras_linhas = df_titanic.iloc[0:3] # Do índice 0 até o 3 (sem incluir o 3)
tres_primeiras_linhas
```

```python
## Selecionando Linhas e Colunas com .loc (Seleção por Rótulo/Label)
# Em nosso DataFrame, o índice padrão são os números das linhas.
# Vamos selecionar as linhas de índice 0 a 4 e apenas as colunas 'Name' e 'Ticket'
df_loc_exemplo = df_titanic.loc[0:4, ['Name', 'Ticket']]
df_loc_exemplo
```

# Filtragem condicional 
Criando uma condição (que retorna True ou False) e usar essa condição para filtrar o DataFrame

```python
## Criando uma condição (Máscara Booleana)
# Esta operação retorna uma Series de True/False
mascara_mulheres = df_titanic['Sex'] == 'female'
mascara_mulheres.head()
```

```python
# Aplicando a Máscara para filtrar
df_mulheres = df_titanic[mascara_mulheres]
df_mulheres.head()
```

```python
## Combinando Múltiplas Condições
# Use & para "E" (ambas as condições devem ser verdadeiras)
# Use | para "OU" (pelo menos uma condição deve ser verdadeira).
# Cada condição deve estar entre parênteses ().

# Pergunta: Quem eram as mulheres da Primeira Classe (Pclass ==1)?
df_mulheres_primeira_classe = df_titanic[(df_titanic['Sex'] == 'female') & (df_titanic['Pclass'] == 1)] 
df_mulheres_primeira_classe.head()
```

```python
# Pergunta: Quem eram as crianças (menores de 10 anos) OU os idosos (maiores de 65 anos)?

df_criancas_idosos = df_titanic[(df_titanic['Age'] < 10) | (df_titanic['Age'] > 65)]
df_criancas_idosos.head()
```

## Criando Novas Colunas
Vamos criar uma coluna Age_in_moths (Idade em Meses)

```python
# Simplesmente atribuímos o resultado de uma operação a uma nova coluna
df_titanic['Age_in_Months'] = df_titanic['Age'] * 12

# Vamos verificar se a coluna foi criada, selecionando-a junto com a original
df_titanic[['Name', 'Age', 'Age_in_Months']].head()
```
